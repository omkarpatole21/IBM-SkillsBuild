### Most booked accommodation type - Single, Couple, Family

```python
# Select single, couple, multiple adults and family
df_not_canceled = df[df['is_canceled'] == 0]

# Select single, couple, multiple adults and family
single   = df_not_canceled[(df_not_canceled.adults == 1) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
couple   = df_not_canceled[(df_not_canceled.adults == 2) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
family   = df_not_canceled[df_not_canceled.adults + df_not_canceled.children + df_not_canceled.babies > 2]
```

```python
# Make the list of Category names, and their total percentage
names = ['Single', 'Couple (No Children)', 'Family / Friends']
count = [single.shape[0],couple.shape[0], family.shape[0]]
count_percent = [x/df_not_canceled.shape[0]*100 for x in count]

# Creating a dictionary to make a DataFrame
d = {'names':names,'count_percent':count_percent}
acc_type = pd.DataFrame(data = d)
```

### Overview of reserved room type and assigned room type

```python
df[['is_canceled','reserved_room_type','assigned_room_type']]
```

```python
def rrt_art_same(df):
  '''
  This function returns whether the customer get their preferred reserved room type or not.
  '''
  if df['reserved_room_type'] == df['assigned_room_type']:
      return 'True'
  else:
      return 'False'
```

```python
# Apply to dataframe
df['rrt_art_same'] = df.apply(rrt_art_same,axis=1)

# Check dataframe
df[['is_canceled','reserved_room_type','assigned_room_type','rrt_art_same']]
```
